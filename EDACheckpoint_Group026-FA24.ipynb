# Research Question

Is there a relationship between sleep and physical activity among adults that are overweight? To analyze this relationship, we will be quantifying quality of sleep by measuring the average hours slept each night. We will be quantifying physical activity by measuring an individuals average step count per day.

## Background and Prior Work

There are many factors that influence sleep, whether that be age, alcohol, or stress. It is shared knowledge that exercise plays a role in the quality and quantity of sleep. It is normal for one to feel more sleepy or drained after exercising. When exercising you're increasing both blood flow and heart rate making you more awake and wired. After some time, sleepiness occurs. The reason for that is because as we exercise our energy in the form of ATP is depleted as our muscles contract and there's an increase in neurotransmitters, resulting in muscle fatigue.  <a href="#cite_note-1">[1]</a> We hope to further prove this point by tracking the heart rate of runners and the quality of sleep one gets through their sleep cycle. Of course we understand that there are other factors to consider when discussing what impacts sleep quality like stress and lifestyle, however, we'll focus on the affects of exercise specifically.

There have been many studies done to support the idea of running having positive effects on people's sleep. One study conducted a test with 51 healthy adolescents running regularly in the morning and compared them to control subjects. They examined the electoencepahlographic patterns and psychological functionings of both groups before and after the three week period. By the end of the test the results were that "objective sleep improved (slow-wave sleep increased; sleep onset latency decreased) in the running group compared with the control group".  <a href="#cite_note-2">[2]</a> Supporting the hypothesis that the quality of sleep improves when running. However, this is only accounting for those in the age of adolescence, the mean age of 18.30 years, so it fails to account for any people older than the age of adolescence.

Another study was done to investigate the effects of exercise in those of an older age group. This study was conducted on participants 40 years or older with the mean age being 62 years old. Participants were then a part of a 12-week exercise training program. Data was measured through sleep quality assessments, a cardiopulmonary exercise test, and heart rate variability assessment. At the end of the program the results were that moderate intensity exercise training had a beneficial significance between exercise and cardiac autonomic function.  <a href="#cite_note-3">[3]</a> The study suggests, even those older who exercise also experience better sleep, similar to the benefits observed in the adolescents.

While many studies we have found support the claim that exercise enhances the sleep quality, we aim to be more specific in our findings. Showcasing the correlation of heart rate when running to the sleep cycles, and the improvement in sleep quality. Additionally considering how the level of experience and gender can connect and affect the outcome as well.

# Hypothesis


Our hypothesis for this questions is that there will be a strong positive correlation between physical activity (measured by daily step counts)  and sleep quality (measured by average hours slept per night) among overweight adults. Adults with high step counts per day are more likely to experience longer sleep durations on average and could also experience lower resting heart rates which indicate good or improving cardiovascular health and sleep quality. We think this is because evidence points to the idea that regular physical activity can promote better sleep because of the release of adenosine that occurs during physical activity. Adenosine can result in a lower resting heart rate and can lead to deeper and higher quality of sleep.
We expect that the positive correlation will be shown through an increase in average hours slept per night, a higher rating for quality of sleep, and a consistent or sustained decrease in individual’s resting heart rate when adults have a consistent or increasing number of daily step counts.

## Amount of Sleep achieved with High/Low Amounts of Steps 

After gathering the Top 10 participants with the highest average number of steps per day and the Bottom 10 participants with the lowest average number of steps per day, we assign them with labels to make a new categorical variable `group` where the two values are `Top 10 (Highest avg_steps)` and `Bottom 10 (Lowest avg_steps)`. Using this variable, we can break down the average minutes of sleep per night by the participants with the most and least steps to observe the relationship between amount of sleep and number of steps.

```python
top_10_df['group'] = 'Top 10 (Highest avg_steps)'
bottom_10_df['group'] = 'Bottom 10 (Lowest avg_steps)'

comparison_df = pd.concat([top_10_df[['avg_sleep', 'group']], bottom_10_df[['avg_sleep', 'group']]])

plt.figure(figsize=(8, 6))
sns.boxplot(x='group', y='avg_sleep', data=comparison_df, palette='Set2')

plt.title('Comparison of avg_sleep between Top 10 and Bottom 10 avg_steps', fontsize=16)
plt.xlabel('Group', fontsize=14)
plt.ylabel('Average Sleep (minutes)', fontsize=14)

plt.tight_layout()
plt.show()
```

The median lines of the boxplots reveal that the group of participants with the lowest average steps per day typically gets 40 more minutes of sleep per night than participants with the highest average steps per day. When comparing the two boxplots, can observe that the group of participants with the highest average steps per day are skewed to have less sleep per night, while the participants with the lowest average steps per day are skewed to have more sleep per night. 

To test the difference in means between the groups, we use a t-test to check for statistical significance.

```python
print(f"Top 10 group mean: {top_10_df['avg_sleep'].mean()}")
print(f"Bottom 10 group mean: {bottom_10_df['avg_sleep'].mean()}") 

print
t_stat, p_value = ttest_ind(top_10_df['avg_sleep'], bottom_10_df['avg_sleep'])

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

#### Interpretation:

Since $p = 0.409$ is much larger than $0.05$, we observe that the difference in means is likely due to random chance alone. This goes against our hypothesis that adults with higher step counts per day are more likely to experience longer sleep durations on average.

## Relationship between Study Length and Missing Data on Sleep, Steps, and Sleep Efficiency

```python
# Compute weights for sleep and steps
df['weight_sleep'] = df['perc_NA_sleep'] * df['study_length']
df['weight_steps'] = df['perc_NA_steps'] * df['study_length']

# Normalize the weights
df['weight_sleep'] /= df['weight_sleep'].sum()
df['weight_steps'] /= df['weight_steps'].sum()
```

```python
# Calculate weighted averages for steps, sleep, and sleep efficiency
weighted_avg_steps = (df['avg_steps'] * df['weight_steps']).sum()
weighted_avg_sleep = (df['avg_sleep'] * df['weight_sleep']).sum()
weighted_avg_sleep_eff = (df['avg_sleep_eff'] * df['weight_sleep']).sum()

print(f"Weighted Average Steps: {weighted_avg_steps:.2f}")
print(f"Weighted Average Sleep: {weighted_avg_sleep:.2f} minutes")
print(f"Weighted Average Sleep Efficiency: {weighted_avg_sleep_eff:.2f}")
```

```python
# Calculate unweighted averages
unweighted_avg_steps = df['avg_steps'].mean()
unweighted_avg_sleep = df['avg_sleep'].mean()
unweighted_avg_sleep_eff = df['avg_sleep_eff'].mean()

# Print comparisons
print("\nComparison of Weighted vs. Unweighted Averages:")
print(f"Steps - Unweighted: {unweighted_avg_steps:.2f}, Weighted: {weighted_avg_steps:.2f}")
print(f"Sleep (minutes) - Unweighted: {unweighted_avg_sleep:.2f}, Weighted: {weighted_avg_sleep:.2f}")
print(f"Sleep Efficiency - Unweighted: {unweighted_avg_sleep_eff:.2f}, Weighted: {weighted_avg_sleep_eff:.2f}")
```

```python
import matplotlib.pyplot as plt

# Bar plot to compare unweighted and weighted averages
labels = ['Steps', 'Sleep (minutes)', 'Sleep Efficiency']
unweighted = [unweighted_avg_steps, unweighted_avg_sleep, unweighted_avg_sleep_eff]
weighted = [weighted_avg_steps, weighted_avg_sleep, weighted_avg_sleep_eff]

x = range(len(labels))
plt.bar(x, unweighted, width=0.4, label='Unweighted', color='blue', alpha=0.7)
plt.bar([i + 0.4 for i in x], weighted, width=0.4, label='Weighted', color='green', alpha=0.7)

plt.xticks([i + 0.2 for i in x], labels)
plt.ylabel('Average')
plt.title('Comparison of Weighted vs. Unweighted Averages')
plt.legend()
plt.show()
```

```python
correlation_sleep = df['perc_NA_sleep'].corr(df['avg_sleep'])
correlation_steps = df['perc_NA_steps'].corr(df['avg_steps'])

print(f"Correlation between Missing Sleep Data (%) and Average Sleep: {correlation_sleep:.2f}")
print(f"Correlation between Missing Steps Data (%) and Average Steps: {correlation_steps:.2f}")
```

```python
perc_NA_sleep = df[df['perc_NA_sleep'] == 1]
high_missing_sleep = df[df['perc_NA_sleep'] != 1]

perc_NA_steps = df[df['perc_NA_steps'] == 1]
high_missing_steps = df[df['perc_NA_steps'] != 1]

# T-test for 'avg_sleep' based on missing data group for sleep
t_stat_sleep, p_val_sleep = ttest_ind(perc_NA_sleep['avg_sleep'], high_missing_sleep['avg_sleep'])

# T-test for 'avg_steps' based on missing data group for steps
t_stat_steps, p_val_steps = ttest_ind(perc_NA_steps['avg_steps'], high_missing_steps['avg_steps'])

print(f"T-test for Average Sleep based on Missing Sleep Data: t-stat={t_stat_sleep:.2f}, p-value={p_val_sleep:.4f}")
print(f"T-test for Average Steps based on Missing Steps Data: t-stat={t_stat_steps:.2f}, p-value={p_val_steps:.4f}")
```

```python
if p_val_sleep < 0.05:
    print(f"There is a significant difference in avg_sleep between the high and low missing data groups (p-value: {p_val_sleep:.4f}).")
else:
    print(f"There is no significant difference in avg_sleep between the high and low missing data groups (p-value: {p_val_sleep:.4f}).")

if p_val_steps < 0.05:
    print(f"There is a significant difference in avg_steps between the high and low missing data groups (p-value: {p_val_steps:.4f}).")
else:
    print(f"There is no significant difference in avg_steps between the high and low missing data groups (p-value: {p_val_steps:.4f}).")
```
